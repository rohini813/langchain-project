- Prompt management
- Chaining multiple operations
- Adding memory to conversations

## Memory

Memory allows the model to remember past interactions.# New Section

```python
chat_history = []

def chat(user_input):
    chat_history.append(f"User: {user_input}")
    context = "\n".join(chat_history) + "\nAI:"
    response = llm.invoke(context)
    response = response.split("AI:")[-1].strip()
    chat_history.append(f"AI: {response}")
    return response

print(chat("Hi, my name is Shyam"))
print(chat("What is my name?"))
```

## Agents

Agents decide whether to use LLM or tools.# New Section

```python
def simple_agent(query):
    if "calculate" in query:
        return "Using calculator tool"
    else:
        return llm.invoke(query)

print(simple_agent("Explain AI"))
```
